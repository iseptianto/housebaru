# üè† House Data Exploration
This notebook performs initial data exploration on the `house_data.csv` dataset.

```python
# Import required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv('/content/edit.csv')

# ‚ÑπÔ∏è Display basic information about the dataset
print(f"üìê Dataset shape (rows, columns): {df.shape}")
print("\nüìä Dataset Info:")
print(df.info())
```

```python
# ‚ùì Checking for missing values to identify data quality issues
print("\nüîç Missing values per column:")
print(df.isnull().sum())
```

```python
# üìà Descriptive statistics like mean, std deviation, min, max, etc.
print("\nüìã Statistical Summary:")
print(df.describe())
```

```python
# Calculate the median of the 'Luas Tanah' column
median_luas_tanah = df['Luas Tanah'].median()

# Replace 0 values in 'Luas Tanah' with the median
df['Luas Tanah'] = df['Luas Tanah'].replace(0, median_luas_tanah)

print(f"üîÑ Replaced 0 values in 'Luas Tanah' with the median ({median_luas_tanah}).")

# Display descriptive statistics for 'Luas Tanah' to verify the change
print("\nüìã Statistical Summary for 'Luas Tanah' after replacing 0s:")
print(df['Luas Tanah'].describe())
```

```python
# Calculate the median of the 'Luas Tanah' column
median_luas_tanah = df['Luas Bangunan'].median()

# Replace 0 values in 'Luas Tanah' with the median
df['Luas Bangunan'] = df['Luas Bangunan'].replace(0, median_luas_tanah)

print(f"üîÑ Replaced 0 values in 'Luas Bangunan' with the median ({median_luas_tanah}).")

# Display descriptive statistics for 'Luas Tanah' to verify the change
print("\nüìã Statistical Summary for 'Luas Bangunan' after replacing 0s:")
print(df['Luas Bangunan'].describe())
```

```python
# üí∞ Visualizing the distribution of house Luas Tanah
plt.figure(figsize=(10, 6))
sns.histplot(df['Luas Tanah'], kde=True)
plt.title('House Luas Tanah Distribution')
plt.xlabel('Luas Tanah')
plt.ylabel('Frequency')
plt.show()

print("üìä Plotted house price distribution.")
```

```python
# üîó Checking how features are correlated with each other
plt.figure(figsize=(12, 8))
correlation_matrix = df.select_dtypes(include=np.number).corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Feature Correlation Matrix')
plt.show()

print("üìà Displayed feature correlation heatmap.")
```

```python
# üìç Bar chart to see how many houses exist in each Type
plt.figure(figsize=(10, 6))
sns.countplot(x='Type', data=df)
plt.title('Houses by Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

print("üìç Visualized distribution of houses by location.")
```

```python
# üìç Bar chart to see how many houses exist in each location
plt.figure(figsize=(10, 6))
sns.countplot(x='Provinsi', data=df)
plt.title('Houses by Provinsi')
plt.xlabel('Provinsi')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

print("üìç Visualized distribution of houses by location.")
```

```python
# üìç Bar chart to see how many houses exist in each location
plt.figure(figsize=(10, 6))
sns.countplot(x='KT', data=df)
plt.title('Houses by KT')
plt.xlabel('KT')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

print("üìç Visualized distribution of houses by location.")
```

```python
# üìç Bar chart to see how many houses exist in each location
plt.figure(figsize=(10, 6))
sns.countplot(x='KM', data=df)
plt.title('Houses by Kamar Mandi')
plt.xlabel('Kamar Mandi')
plt.ylabel('Count')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

print("üìç Visualized distribution of houses by location.")
```

```python
# üõèÔ∏è Bar chart showing how average house price varies by bedroom count
plt.figure(figsize=(10, 6))
sns.barplot(x='KT', y='KM', data=df)
plt.title('Average Price by Number of Rooms (KT)')
plt.xlabel('Number of Rooms (KT)')
plt.ylabel('Kamar Mandi')
plt.show()

print("üõèÔ∏è Compared average prices based on bedroom count.")
```
